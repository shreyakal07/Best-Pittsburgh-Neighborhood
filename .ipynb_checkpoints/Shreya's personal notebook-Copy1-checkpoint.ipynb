### Safety in Pittsburgh Neighborhoods 

The last factor we took into consideration was crime rate.

**Metric**

Higher rate = high safety

Lower rate = low safety


```python
crime = pd.read_csv("Data/crime_rates.csv")



crime['INCIDENTNEIGHBORHOOD'].value_counts().sort_values(ascending=True)
```

```python
crime[['INCIDENTNEIGHBORHOOD']].describe()
```

```python
##I determined the metric by using the lowest number of crimes(in this case-it is 1) 
## and dividing it by the number of crimes in each neighborhood. 
##The safety rate of each neighborhood is shown underneath, whereas the higher the rate, the safer it is.
df = (1 / crime['INCIDENTNEIGHBORHOOD'].value_counts()) * 100
df
```

```python
df3 = df.plot(kind = "bar", color = "teal", figsize = (20,10))
df3
```

```python
df.tail(5)
```

```python
#top 5 safest neighborhoods

crimeRate = {'Total Crimes': ['1', '1', '2', '2', '2'], 
        'Safety Rate': [100, 100, 50, 50, 50], 
        'Hierachy': [27, 11, 9, 99, 27]}
df = pd.DataFrame(crimeRate, index = ['St. Clair', 'New Homestead', 'Outside County', 'Oakwood', 'Mt. Oliver Boro'])
df
```

```python
df.head(5)
```

```python
#top 5 worst neighborhoods
crimeRate = {'Total Crimes': ['211', '125', '112', '84', '82'], 
        'Safety Rate': [0.47, 0.8, 0.89, 1.19, 1.22], 
        'Hierachy': [99, 99, 99, 99, 99]}
df = pd.DataFrame(crimeRate, index = ['Central Business District', 'South Side Flats', 'Carrick', 'Homewood South', 'East Liberty'])

df
```

#### Worst Neighborhood ####

- Central Business District
    - The worst neighborhood was determined by looking at the chart above. We see that the total number of crimes in this neighborhood is 211; therefore, the safety rate was roughly 0.47. To make this data a little more fair, I included the highest hierarchy that occured in each neighborhood. All of which had 99, meaning a high chance of potential risks. 

#### Best Neighborhood ####

- New Homestead
    - The best neighborhood had a safety rate of 100. With New Homestead and St. Clair having the same safety rate, I included the hierarchy which showed the ranking of crime for each. The higher the hierarchy value, the more fatal the crime is. For New Homestead, the crime was an access fraud, while for St. Claire the crime was a hit and run. Consequently, New Homestead was the safer choice of the two.
